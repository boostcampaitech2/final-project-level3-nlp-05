```python
import seaborn as sns
import pandas as pd
import json
```

```python
from wikitree_crawling import CrawlingWikitree

crawling_obj = CrawlingWikitree(click_cnt=1)
categories = crawling_obj.categories.keys()
print(categories)
```

```python
data = {
    'category':[],
    'date':[]
}
for category in categories:
    with open(f'article_{category}.json', 'r') as f:
        articles = json.load(f)

    for article in articles:
        data['category'].append(article['category'])
        data['date'].append(article['publish_date'][:10])

df = pd.DataFrame(data)

```

```python
sns.countplot(y='date',data=df.loc[df.category == '경제',:])
```
